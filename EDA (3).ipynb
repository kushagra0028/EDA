```python
import numpy as np # for maths operations
import pandas as pd # for better arrangement of dataset
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('anime.csv')
```

```python
df.shape
# to find number of rows and columns in a dataset
```

```python
# Dataset has 24985 rows and 39 columns (parameters)
```

```python
df.head()
# shows by default 5 rows but if writtern like df.head(n) --> shows n rows
```

```python
df.info()
# used to obtain a concise summary of a DataFrame’s structure and information
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['score'] = df['score'].astype('int')
```

```python
df['score'].dtypes
# changed from float to int
```

```python
df.columns
# shows column name
```

```python
#rename column
df.rename(columns= {'producers':'Animators'})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
# learned by own sir so please don't consider as UFM.
```

```python
# describe can be used for specific columns too
df[['anime_id', 'score', 'episodes']].describe()
```

```python
# plotting a bar chart for score and it's episodes using seaborn

ax = sns.countplot(x = 'score',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total number of anime from top 10
sales_state = df.groupby(['anime_id'], as_index=False)['score'].sum().sort_values(by='score', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)}) # for setting figure size
sns.barplot(data = sales_state, x = 'anime_id',y= 'score')
```

```python
sns.barplot(data=df, x= 'source', y = 'score')
# here we can see Confidence interval lines which are visual representations of a confidence interval on a graph or plot.
# It depict the range of values within which the true population parameter is likely to fall
```

```python
sns.boxplot(y='score', x = 'source' , data = df)
#  used to depict the distribution of data across different quartiles --> reference google
# here the points lying outside are called outliers. These are data points that fall significantly beyond the typical range of the data set.
```

```python
df[(df.type == 'movie') & (df.score >9.5)]
```

```python
# there is no anime in dataset having a movie rating more than 9.5 
```

```python
producerList=[]
for key, value in df.items():
    temp = [key,value]
    producerList.append(temp)
```

```python
producer_df=pd.DataFrame(producerList, columns = ["Producer", "Count"]) 
producer_df
```

```python
producer_top10_df=df.sort_values("score").tail(10)
producer_top10_df
# Finding top 10 animes from dataset from below
```

```python
producer_top10_df=df.sort_values("score").head(10)
producer_top10_df
# Finding top 10 animes from dataset from top
```

```python
plt.title("Top 10 Anime animators")

sns.barplot(x='score', y='producers', data=producer_top10_df)

# Show the plot
plt.show()
```

```python
source_df=df.groupby('source')[['favorites']].mean()
source_df
# query to find favourite source of anime
# .groupby('source'): This part groups the DataFrame df by the source column. This means it creates separate groups for data points with each unique source label
```

```python
#Here I use the plotArea and plotPie function using Matplotlib.
plt.rcParams['figure.figsize'] = (10, 10)
source_df.plot.area()
```

```python
source_df.plot.pie(y="favorites")
```

```python
#recommending some Animes from the dataset to the community by analysis.
rec_anime_df=df.loc[(df['favorites'] > 50000) & (df['score'] > 7)]
rec_anime_df
```
